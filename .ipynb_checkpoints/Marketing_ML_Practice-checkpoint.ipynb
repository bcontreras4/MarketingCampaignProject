```python
# Again, we've converted year each entry became a customer and assigned it a numerical value
# The smaller this new number label,the longer they have been a customer

Dt_le = LabelEncoder()
Dt_Labels = Dt_le.fit_transform(campaign_info["Dt_Customer"])
Dt_mappings = {index: label for index, label in 
                  enumerate(Dt_le.classes_)}

campaign_info["Dt_Labels"] = Dt_Labels
```

### So far, we have our original campaign_info data with added Dt labels

```python
campaign_info.head()
```

```python
campaign_info.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# Creates new category that tells us whether a customer accepted a campaign at any point or not

campaign_info["AcceptedAnyCmp"] = (campaign_info["AcceptedCmp1"] + campaign_info["AcceptedCmp2"] + \
    campaign_info["AcceptedCmp3"] + campaign_info["AcceptedCmp4"] + campaign_info["AcceptedCmp5"]) >= 1

# Turns boolean "True/False" into integers "1/0" so the category can be plotted as a histogram

campaign_info["AcceptedAnyCmp"] = campaign_info["AcceptedAnyCmp"].astype(np.int8)
```

```python
have# Creates a new category that shows the number of campaigns accepted by each customer

campaign_info["AcceptedNumofCmps"] = (campaign_info["AcceptedCmp1"] + campaign_info["AcceptedCmp2"] + \
    campaign_info["AcceptedCmp3"] + campaign_info["AcceptedCmp4"] + campaign_info["AcceptedCmp5"])
```

### Now, we have or modified campaign_info with new categories of "AcceptedAnyCmp" and "AcceptedNumofCmps"

```python
campaign_info.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# We now create the test and training sets; We stratify these based on "AcceptedAnyCmp" and "Recency"

split = StratifiedShuffleSplit(n_splits = 1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(campaign_info, campaign_info["AcceptedAnyCmp"]):
    strat_train_set = campaign_info.loc[train_index]
    strat_test_set = campaign_info.loc[test_index]
```

```python
# We are getting rid of the 7 customers who had nonsensicle answers for their marital status

strat_train_set = strat_train_set[strat_train_set["Marital_Status"] != "Alone"]
strat_train_set = strat_train_set[strat_train_set["Marital_Status"] != "Absurd"]
strat_train_set = strat_train_set[strat_train_set["Marital_Status"] != "YOLO"]

strat_test_set = strat_test_set[strat_test_set["Marital_Status"] != "Alone"]
strat_test_set = strat_test_set[strat_test_set["Marital_Status"] != "Absurd"]
strat_test_set = strat_test_set[strat_test_set["Marital_Status"] != "YOLO"]
```

```python
''''
MS_le = LabelEncoder()
MS_Labels = MS_le.fit_transform(strat_train_set["Marital_Status"])
MS_mappings = {index: label for index, label in 
                  enumerate(MS_le.classes_)}

strat_train_set["MS_Labels"] = MS_Labels

MS_mappings

MS_le = LabelEncoder()
MS_Labels = MS_le.fit_transform(strat_test_set["Marital_Status"])
MS_mappings = {index: label for index, label in 
                  enumerate(MS_le.classes_)}

strat_test_set["MS_Labels"] = MS_Labels
'''

# UPDATE: Now using OneHotEncoder instead of the MS_Labels so that the ML algorithm does not confuse label
# proximity with label similarity
```

### At this point, we have 1 training set and one test set. Both have Education labels, Dt labels, AcceptedNumCmps, AcceptedAnyCmps, and we've added Marital Status Labels to both while dropping any rows with MS answers "Absurd/Alone/YOLO" from BOTH sets.

```python
# Looking at how well the test set was stratified. We would expect ~80% of test set to be 0 and remainder 1

strat_test_set["AcceptedAnyCmp"].value_counts() / len(strat_test_set)
```

```python
campaign_info = strat_train_set.copy()

for set_ in (campaign_info, strat_test_set):
    set_.drop("ID", axis=1, inplace=True)
```

### Here, our test set is the same as above, but we are now working with a COPY of the training set. Both the test set and this copy have the ID column dropped from them.

```python
# First, we must only include numerical values in our corr_matrix so we'll get a numerical version of our data

obj_attributes = ["Dt_Customer", "Education", "Marital_Status"]
num_campaign_info = campaign_info.drop(obj_attributes, axis=1)

# We also do not get any correlation info from Z_Revenue & Z_CostContact, so I will remove those too
Z_attributes = ["Z_Revenue", "Z_CostContact"]
num_campaign_info = num_campaign_info.drop(Z_attributes, axis=1)

corr_matrix = num_campaign_info.corr()
corr_matrix["MntWines"].sort_values(ascending=False)
fig = plt.figure(figsize = (10,10))
plt.matshow(corr_matrix, 0, cmap='inferno')

# Below, we set up the matrix axis label by getting the column names and number of columns
plt.xticks(np.linspace(0, len(num_campaign_info.columns)-1, len(num_campaign_info.columns)), 
           num_campaign_info.columns, rotation=90)
plt.yticks(np.linspace(0, len(num_campaign_info.columns)-1, len(num_campaign_info.columns)), 
           num_campaign_info.columns, rotation=0)
corr_norm = colors.Normalize(vmin=-1,vmax=1)
plt.colorbar(cm.ScalarMappable(norm=corr_norm, cmap='inferno'))
```

 So far, the training set shows that income plays a small role in how much a costomer spends on products, which is expected. Intrestingly, it has a strong negative corrolation with number of web visits per month. In fact, many attributes had a slight negative correlation with web visits per month except the number of deals purchases and kids at home. If a customer had kids at home there was less likelyhood that they would accept any of the campaigns and while they made more web visits per month, they were not making more web purchases and spent less on products over all. This effect was not seen so much in those with teens at home. So, it seems many of the customers are using the store and catalog to purchase products, especially the ones that spend more. In fact, the more they spent on any of the products, the less their web visits were. It also seems that wine buyers were more likely to accept a campaign, in particular campaign 5 was successful with this group the more they spent and not campaign 3. The more campaigns a customer accepted, the more likely they agreed to the last response. Big meat products spenders are often buying from catalog and this group also had a slight positive correlation with whether they accepted campaign 5 or not.

### At this point we have our modified test set and our copy of the training set. We've also gotten rid of Z_revenue and Z_CostContact since they do not have any significant effect on the data and we are not sure of their meaning at this point in time
